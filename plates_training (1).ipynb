```python
import os

# Use raw strings (r"") to prevent escape character issues
dataset_path = r"/content/drive/MyDrive/darknet/data/Label_new"
train_txt_path = r"/content/drive/MyDrive/darknet/data/valid.txt"
# Get all image files (modify extensions if needed)
image_extensions = (".jpg", ".png", ".jpeg")
image_files = [f for f in os.listdir(dataset_path) if f.endswith(image_extensions)]

# Write full paths to train.txt
with open(train_txt_path, "w") as f:
    for image in image_files:
        full_path = os.path.join(dataset_path, image)
        f.write(full_path + "\n")

print(f"✅ valid.txt created successfully with {len(image_files)} entries!")
```
